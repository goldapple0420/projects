```python
key = []

total = requests.get("https://asms.coa.gov.tw/Asms/api/ViewNowAnimal?pageSize=10000&currentPage=1&sortDirection=DESC&sortFields=AcceptDate").text
totalJ = json.loads(total)

for i in totalJ:
    keyDict = {}
    keyDict["Id"] = i["AnimalId"]
    keyDict["AcceptNum"] = i["AcceptNum"]
    key.append(keyDict)

        
```

```python
ua = UserAgent()
```

```python
Ainfo = []

for i in key:
    try:
        Pinfo = {}
        user_agent = ua.random
        headers = {'user-agent': user_agent}
        info = requests.get("https://asms.coa.gov.tw/Asms/api/Animals?AcceptNum={}&keyNo={}&stChk=Y".format(i['AcceptNum'],i['Id']), headers=headers).text
        infoJ = json.loads(info)[0]
        exist("Name")
        exist("TypeIdName")
        exist("SexName")
        exist("VarietyName")
        exist("AcceptDate")
        exist("Age2Name")
        exist("ShelterName")
        exist("CoatName")
        exist("BuildName")
        exist("IsTNR")
        exist("HealthStatusName")
        Pinfo["city"] = Pinfo["ShelterName"][0:3]
        try:
            Pinfo["image_site"] = "https://asms.coa.gov.tw/Amlapp/Upload/Pic/" + infoJ["InImage"]
        except:
            Pinfo["image_site"] = "未提供"
        try:    
            Pinfo["url"] = "https://asms.coa.gov.tw/Amlapp/App/AnnounceList.aspx?Id={}&AcceptNum={}&PageType=Adopt".format(i['Id'],i['AcceptNum'])
        except:
            Pinfo["url"] = "未提供"
        Ainfo.append(Pinfo)
    except:
        print(i)
        continue
    time.sleep(3)
```
